```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('graph.csv')
df['datetime']=pd.to_datetime(df['datetime'])
df=df.set_index('datetime')
```

```python
fig, ax1 = plt.subplots(figsize=(16,8))
ax2 = ax1.twinx()
ax3 = ax1.twinx()
ax1.bar(df.index,df['sell_volume'],0.0005,label='sell_volume')
ax1.bar(df.index,df['buy_volume'],0.0005,label='buy_volume')


ax2.plot(df['price'],label='price')
ax2.plot(df['ema0'],label='ema0')
ax2.plot(df['ema1'],linestyle='dashed',label='ema1')
ax3.plot(df['asset'],color='r',label='asset')


ax1.legend(bbox_to_anchor=(1, 1), loc='lower right', borderaxespad=0, fontsize=10)
ax2.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=2, fontsize=10)
ax3.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=10)
```
